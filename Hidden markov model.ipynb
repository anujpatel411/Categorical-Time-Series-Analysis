```python
import sys

np.set_printoptions(threshold=sys.maxsize)
import urllib
import requests
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import multinomial
from typing import List
import seaborn as sns
from matplotlib.ticker import FormatStrFormatter
import random
from hmmlearn import hmm
from array import array
```

```python
link = "https://www.ebi.ac.uk/ena/browser/api/fasta/CP009973.1?lineLimit=1000"
f = requests.get(link)


data = f.text
```

```python
data
```

```python
str = '>ENA|CP009973|CP009973.1 Yersinia pestis CO92, complete genome.\n'
```

```python
data = data[64:len(data)]
```

```python
data = data.replace('\n','')
data = data.replace('G','1')
data = data.replace('A','2')
data = data.replace('C','3')
data = data.replace('T','4')
data = data.replace(' ','' )
```

```python
data
```

```python
data1 = []
for i in data:
    data1.append(int(i))
```

```python
len(data1)
```

```python
model = hmm.CategoricalHMM(n_components=4, random_state = 0)
```

```python
emissions = np.array(data1).reshape(-1, 1)
emissions
```

```python
model.fit(emissions)
```

```python
hidden_states = model.predict(emissions)
hidden_states
```

```python
trans_mat = model.transmat_
trans_mat
```

```python
emission_prob=model.emissionprob_
emission_prob
```

```python
start_prob=model.startprob_
start_prob
```
